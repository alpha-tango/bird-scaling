```python
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
```

```python
landmass_areas = gpd.read_file('outputs/shapefiles/albers_antarctic_areas.shp')
```

```python
landmass_areas.shape
```

```python
biggest_areas = landmass_areas.sort_values('area_km2', ascending=False).head(10)
biggest_areas
```

```python
colors = ['tab:blue', 'tab:green', 'tab:gray', 'tab:orange', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:olive', 'tab:cyan']
biggest_areas.plot(color=colors)
```

```python
landmass_areas.sort_values('area_km2', ascending=True).head(10)
```

```python
landmass_areas['log_area'] = np.log(landmass_areas['area_km2']).round()
fig, ax = plt.subplots()
ax.hist(landmass_areas['log_area'])
```
