### OSM Open Streetmap Coastline Data

[background info](https://osmdata.openstreetmap.de/processing/coastline.html)
[additional details](https://wiki.openstreetmap.org/wiki/Tag:natural=coastline)
"Those ways need to connect end-to-end to form an unbroken line around every island and every continent. And the land always has to be on the left side, the water on the right side of those ways."

NOT used for lakes (except the Caspian Sea) and rivers!

And, it's made using the high water mark:
"Mean high water springs (MHWS) is the averaged highest level that spring tides reach over many years (often the last 19 years). Within this, to ensure anomalous levels are tempered, at least two successive high waters during the highest-tide 24 hours are taken.[16]

Such a local level is generally close to the "high water mark" where debris accumulates on a tidal shore on about two days six months apart (and nearby days) annually. The levels are local as some places are nearer to or form places of almost no tides in and around each ocean (amphidromic points)."
